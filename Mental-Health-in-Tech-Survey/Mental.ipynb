```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import plotly.express as px
sns.set_style('whitegrid')
%matplotlib inline
```

```python
df = pd.read_csv('survey.csv')
```

```python
df.head()
```

```python
df.columns = df.columns.str.lower()
```

**timestamp** - Отметка времени

**age** - Возраст

**gender** - Пол

**country** - Страна

**state** - если вы живете в Соединенных Штатах, в каком штате или территории вы живете?

**self_employed** - Вы работаете не по найму?

**family_history** - У вас есть семейная история психических заболеваний?

**treatment** - Обращались ли вы за лечением по поводу психического расстройства?

**work_interfere** - Если у вас есть проблемы с психическим здоровьем, считаете ли вы, что они мешают вашей работе?

**no_employees** - Сколько сотрудников в вашей компании или организации?

**remote_work** - Работаете ли вы удаленно (вне офиса) не менее 50% времени?

**tech_company** - Является ли ваш работодатель преимущественно технологической компанией/организацией?

**benefits** - Предоставляет ли ваш работодатель льготы по охране психического здоровья?

**care_options** - Знаете ли вы, какие варианты психиатрической помощи предоставляет ваш работодатель?

**wellness_program** - Ваш работодатель когда-нибудь обсуждал психическое здоровье в рамках программы оздоровления сотрудников?

**seek_help** - Предоставляет ли ваш работодатель ресурсы, чтобы узнать больше о проблемах психического здоровья и о том, как обратиться за помощью?

**anonymity** - Защищена ли ваша анонимность, если вы решите воспользоваться ресурсами для лечения психического здоровья или наркомании?

**leave** - Насколько легко вам взять отпуск по болезни в связи с психическим заболеванием?

**mental_health_consequence** - Считаете ли вы, что обсуждение проблемы психического здоровья с вашим работодателем может иметь негативные последствия?

**phys_health_consequence** - Считаете ли вы, что обсуждение проблемы физического здоровья с вашим работодателем может иметь негативные последствия?

**coworkers** - Хотели бы вы обсудить проблему психического здоровья со своими коллегами?

**supervisor** - Не хотели бы вы обсудить проблему психического здоровья со своим непосредственным руководителем?

**mental_health_interview** - Поднимете ли вы проблему психического здоровья потенциальному работодателю на собеседовании?

**phys_health_interview** - Поднимете ли вы вопрос физического здоровья потенциальному работодателю на собеседовании?

**mental_vs_physical** - Считаете ли вы, что ваш работодатель относится к психическому здоровью так же серьезно, как и к физическому?

**obs_consequence** - Слышали ли вы или наблюдали ли вы о негативных последствиях для коллег с психическими расстройствами на вашем рабочем месте?

**comments** - Любые дополнительные примечания или комментарии
