```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('C:/Users/aliso/spotify/tracks.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df_corr = df[['popularity','duration_ms','danceability','energy','loudness','speechiness','acousticness','instrumentalness','liveness','valence','tempo']]
```

```python
df_corr.head()
```

```python
corr = df_corr.corr()

f,ax = plt.subplots(figsize=(9,6))
sns.heatmap(corr, annot = True, linewidths=1.5 , fmt = '.2f',ax=ax)
plt.show()
```

```python
df_90 = df.query('popularity >= 90').sort_values('popularity', ascending=False)
```

# Громкость песни напрямую влияет на ее популярность. Чем громче песня, тем она популярнее

```python
plt.figure(figsize=(22,10))
ax = sns.barplot(x='popularity', y='energy', data=df, color='blue', alpha=0.5)
ax.set(xlabel='Popularity', ylabel='energy')
sns.despine()  # убрать часть рамки графика
plt.show()
```

```python
plt.figure(figsize=(22,10))
ax = sns.barplot(x='popularity', y='danceability', data=df, color='blue', alpha=0.5)
ax.set(xlabel='Popularity', ylabel='danceability')
sns.despine()  # убрать часть рамки графика
plt.show()
```

```python
plt.figure(figsize=(22,10))
ax = sns.barplot(x='popularity', y='acousticness', data=df, color='blue', alpha=0.5)
ax.set(xlabel='Popularity', ylabel='acousticness')
sns.despine()  # убрать часть рамки графика
plt.show()
```

## Более популярные песни длятся меньше. Энергичность, танцевальность песни влияет на ее популярность. Акустичность песни обратно пропорциональна ее популярности, популярные песни менее акустична
    

```python
df.head()
```

```python
df.describe()
```

```python
df['release_date'] = pd.to_datetime(df['release_date'])
```

```python
df['year'] = df['release_date'].dt.year
```

```python
df.head()
```

```python
duration_by_year = df.groupby('year', as_index=False).agg({'duration':'mean'})
```

```python
plt.figure(figsize=(16,8))
sns.lineplot(x='year', y='duration', data=duration_by_year)
```

```python
loudness_by_year = df.groupby('year', as_index=False).agg({'loudness':'mean'})
```

```python
plt.figure(figsize=(16,8))
sns.lineplot(x='year', y='loudness', data=loudness_by_year)
```

```python
tempo_by_year = df.groupby('year', as_index=False).agg({'tempo':'mean'})
```

```python
plt.figure(figsize=(16,8))
sns.lineplot(x='year', y='tempo', data=tempo_by_year)
```

```python
popularity_by_year = df.groupby('year', as_index=False).agg({'popularity':'mean'})
```

```python
plt.figure(figsize=(16,8))
sns.lineplot(x='year', y='popularity', data=popularity_by_year)
```
