```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('C:/Users/aliso/Projects/Pet-project/Games-analysis/vgsales.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.duplicated().sum()
```

```python
df.shape
```

```python
df = df.loc[~df['Year'].isna()]
```

```python
df['Year']=df['Year'].astype('int')
```

```python
df = df.drop(columns=['Rank'])
```

```python
df.head()
```

```python
df.groupby(['Year'], as_index=False) \
    .agg({'NA_Sales':'sum','EU_Sales':'sum','JP_Sales':'sum','Other_Sales':'sum','Global_Sales':'sum'}) \
    .sort_values('Year', ascending = False)
```

```python
df.groupby(['Platform'], as_index=False) \
    .agg({'Global_Sales':'sum'}) \
    .sort_values('Global_Sales', ascending=False)
```

```python
plt.figure(figsize=(12,6))
ax = sns.countplot(x="Year", data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90);
ax.set_title('Продажи в период с 1980 по 2016 гг.');
```

```python
df.groupby(['Platform']) \
    .agg({'Global_Sales':'sum'}) \
    .sort_values('Global_Sales', ascending=False) \
    .plot(kind='bar', figsize=(12,6), title='Продажи по платформам в период с 1980 по 2016 гг.');
```

```python
df06 = df.query('Year >= 2006')
```

```python
df06.head()
```

```python
df06.groupby(['Platform']) \
    .agg({'Global_Sales':'sum'}) \
    .sort_values('Global_Sales', ascending=False) \
    .head(10) \
    .plot(kind='bar', figsize=(12,6), title = 'Продажи по платформам в период с 2006 по 2016 гг.');
```

```python
df06.groupby(['Platform']) \
    .agg({'NA_Sales':'sum', 'EU_Sales':'sum', 'JP_Sales':'sum'}) \
    .sort_values('NA_Sales', ascending=False) \
    .head(10) \
    .plot(kind='bar', figsize=(12,6), title = 'Продажи по платформам по регионам в период с 2006 по 2016 гг.');
```

```python
df06.groupby(['Platform']) \
    .agg({'Global_Sales':'sum'}) \
    .sort_values('Global_Sales', ascending=False) \
    .head(10) 
```

```python
a = df06.pivot_table(index='Year', columns='Platform', values='Global_Sales', aggfunc='sum')
a.plot(figsize=(20,10), grid=True, title='Продажи в период с 2006 по 2016 гг.')
plt.xlim(2006,2017,)
plt.xlabel('Год')
plt.ylabel('Сумма продаж(млн)')
plt.show()
```

```python
a.plot(figsize=(20,10), grid=True, title='Продажи в период с 2012 по 2016 гг.')
plt.xlim(2012,2017,)
plt.xlabel('Год')
plt.ylabel('Сумма продаж(млн)')
plt.show()
```
