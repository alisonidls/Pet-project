```python
axes = df.drop(['year', 'duration'], axis=1)\
         .hist(figsize=(16,12), bins=20, alpha=0.75);
axes = axes.ravel()
plt.tight_layout()

titles = ['Popularity', 'Explicit', 'Release Date',
          'Danceability', 'Energy', 'Key', 'Loudness', 'Mode', 'Speechiness',
          'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo', 
          'Time Signature', 'Duration (Min)']
for i in range(len(axes)):
    axes[i].set_title(titles[i])
    if i % 4 == 0:
        axes[i].set_ylabel('Number of Songs')
plt.suptitle('Distributions of Spotify Audio Features', fontsize=30, y=1.05);
```
