```python
df=df.drop(columns='Unnamed: 0')
```

```python
def to_clean(val):
    return val.strip().lower().replace(" ","_")
df.rename(columns=to_clean, inplace= True)
```

```python
df.rename(columns={'averageprice':'average_price'},inplace=True)
```

```python
df['date']=pd.to_datetime(df['date'],format='%Y-%m-%d')
```

```python
df['month']=df['date'].dt.month
df['day']=df['date'].dt.day
```

```python
df.region.unique()
```

```python
regions=['Midsouth','Northeast','SouthCentral','Southeast','West','TotalUS']
```

```python
df_regions = df.query('region == @regions')
```

```python
df=df.query('region !=@regions')
```

```python
df_regions.head()
```

```python
plt.figure(figsize=(16,6))
ax= sns.lineplot(data=df, x="date", y="total_volume")
ax.yaxis.grid(True)
ax.set_title('Total volume of sold avocado')
```

```python
plt.figure(figsize=(16,6))
ax= sns.lineplot(data=df, x="date", y="average_price")
ax.yaxis.grid(True)
ax.set_title('Average price of sold avocado')
```

```python
df_avg_price=df.groupby(['year','month'],as_index=False) \
    .agg({'average_price':'mean'})
```

```python
df_avg_price['y-m'] = df_avg_price['year'].astype(str)+"-"+df_avg_price['month'].astype(str)
df_avg_price['y-m']=pd.to_datetime(df_avg_price['y-m'],format='%Y-%m')
```

```python
plt.figure(figsize=(16,6))
ax= sns.lineplot(data=df_avg_price, x="y-m", y="average_price")
ax.yaxis.grid(True)
ax.set_title('Average price of sold avocado per month');
```

```python
df_states=df.groupby('region',as_index=False) \
    .agg({'average_price':'mean','total_volume':'sum'})
```

```python
df_states=df_states.round(2)
```

```python
df_states.head()
```

```python
plt.figure(figsize=(15,15))
ax = sns.barplot(x='average_price', y='region',
                 data=df_states,palette=['#72b600'])
ax.bar_label(ax.containers[0]);
```

```python
plt.figure(figsize=(15,15))
ax = sns.barplot(x='total_volume', y='region',
                 data=df_states,palette=['#72b600'])
ax.bar_label(ax.containers[0]);
```
